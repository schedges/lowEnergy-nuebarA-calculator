```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

fname1="outputs/1H_total.csv"
fname2="outputs/63Cu_total.csv"

def loadCSVFile(name):
  f = open(name,"r")
  xs=[]
  ys=[]
  for line in f:
    if not line=="":
      line=line.strip("\n")
      line=line.strip()
      lineParts=line.split(",")
      xs.append(float(lineParts[0]))
      ys.append(float(lineParts[1]))
  return xs,ys

ibd_xs,ibd_ys=loadCSVFile(fname1)
cu_xs,cu_ys=loadCSVFile(fname2)

fig,ax=plt.subplots()
plt.plot(ibd_xs,ibd_ys,label="$^1$H IBD")
plt.plot(cu_xs,cu_ys,label="$^{63}$Cu IBD")
plt.yscale("log")
plt.xlabel("Neutrino energy (Mev)")
plt.ylabel("Cross section (cm$^2$)")
plt.legend()
plt.ylim(1e-48,1e-41)
plt.xticks([1*i for i in range(0,11)])
ax.xaxis.set_minor_locator(AutoMinorLocator())
plt.xlim(0,5)
plt.savefig("plots/IBD_63Cu_comp.png",dpi=400)
plt.show()
```
